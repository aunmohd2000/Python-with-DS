- selecting rows and columns
   - rows(slicing)
     - 'loc'     - gets us the data using named indexes
     - 'iloc'    - gets us the data using numeric indexes
   - colms (dictionary style)
        - df[single] - requires name of column
        - df[multiple] - requires a list of columns(multiple is list of named cods)

```python
# single row (record)
weather.iloc(0)
```

```python
weather.iloc[-1]
```

```python
# records  1-50
weather.iloc[25:50]
```

```python
# records 1 to 15 with first 5 colms
weather.iloc[1:16, :5]
```

```python
# records 1 to 15 with first 5 c]olms
weather.iloc[1:16,[0,1,2,3,4]]
```

```python
# 15 records with last 3 colms
weather.iloc[:15, [-3,-2,-1]]
```
